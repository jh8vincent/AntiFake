### 3. Validating of the Suburb Name with Legal Suburb Name

```python
# Generate the legal suburb name in metro melbourne
melbourne_suburb = melbourne_suburb.drop(columns=['id', 'neeed to be changed', 'changed'])
property_suburb['region'] = property_suburb['region'].str.upper().str.strip()
property_suburb = property_suburb.dropna(subset=['region'])
# Display the result
melbourne_suburb
```

```python
# Validate the number of suburb name in the given dataset
prop_suburb_list = list(property_suburb['region'])
len(prop_suburb_list)

melbourne_suburb_list = list(melbourne_suburb['suburb'])
len(melbourne_suburb_list)
```

```python
# Find the identical name list
same_list = []

for i in melbourne_suburb_list:
    for x in prop_suburb_list:
        if x == i:
            same_list.append(x)
            
len(same_list)
```

```python
# Generate the illegal suburb name list
list(set(prop_suburb_list) - set(same_list))
```

### 5.Genrating the top 5 Overall Reliability Property

```python
# Rename and drop the duplicate column
property_suburb = property_suburb.drop(['id'], axis=1)
property_suburb = property_suburb.rename(columns={'name_x': "name", 'picture_url_x': "picture_url",
                                'latitude_y': "latitude", 'region': 'suburb'})
property_suburb.shape
```

```python
# Display the updated dataframe
property_suburb['neighbourhood_cleansed'] = property_suburb['neighbourhood_cleansed'].str.upper().str.strip()
property_suburb
```

```python
# Generate the updated property suburb file
property_suburb.to_csv('property_suburb.csv', index=False)
```

```python
# Generate the top 5 Overall Reliability Property for each suburb
suburb_top5 = property_suburb.sort_values('Overall_Reliable_probability',ascending = False).groupby('suburb').head(5).reset_index()
suburb_top5.shape
```

```python
# Generate the output file
suburb_top5 = suburb_top5.drop(['index'], axis=1)
melbourne_suburb.to_csv('Suburb_Info.csv', index=True)
suburb_top5.to_csv('Top5_Property_by_Suburb.csv', index=True)
```
